# 구구단 출력하기
> 2단부터 9단까지 반복문을 사용하여 구구단을 출력하세요.
---
```
예시 출력)
------- [2 단] -------
2 X 1 = 2
2 X 2 = 4
2 X 3 = 6
2 X 4 = 8
2 X 5 = 10
2 X 6 = 12
2 X 7 = 14
2 X 8 = 16
2 X 9 = 18

------- [3 단] -------
...
```

````python
# 아래에 코드를 작성하세요.
````

````python
for dan in range(2, 10):
    print(f'------- [{dan} 단] -------')
    for multi in range(1, 10):
        print(f'{dan} X {multi} = {dan * multi}')
    print()
````

# 달력 출력하기

> 1월 1일 월요일부터 12월 31일까지 달력을 출력하세요.

---
```

```

````python
calendar = { 
        1: 31, 2: 28, 3: 31, 4: 30, 5: 31,6: 30, 
        7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31 
} 
weeks = ['Mo', 'Tu', 'We', 'Th', 'Fr', 'Sa', 'Su'] 

# 아래에 코드를 작성하세요.
````

````python
for month, day in calendar.items():
    counter = 0
    print(f'         {month} 월')
    print(' '.join(weeks))
    for each in range(1, day + 1):
        if each < 10:
            each = ' ' + str(each)
        print(f'{each}', end = ' ')
        if counter % 7 == 6:
            print(end = '\n')
        counter += 1
    print('\n')
````

# 개인정보보호
> 사용자의 핸드폰번호를 입력 받고, 개인정보 보호를 위하여 뒷자리 4자리를 제외하고는 마스킹 처리하세요.
> * 핸드폰번호는 010으로 시작해야하고 11자리여야한다.
> * 핸드폰번호를 입력하지 않았다면 "핸드폰번호를 입력하세요"를 출력한다.

---

```
예시 입력)
01012341234

예시 출력)
*******1234
```

````python
# 아래에 코드를 작성하세요.
given = input()

if (not given) or len(given) != 11 or given[:3] != '010':
    print('핸드폰번호를 입력하세요')
else:
    print('*' * (len(given) - 4) + given[7:])
````

# 정중앙
> 사용자가 입력한 문자열중 가운데 글자를 출력하세요. 
>
> 단, 문자열이 짝수라면 가운데 두글자를 출력하세요.

---

```
예시 입력)
abc

예시 출력)
b
```

````python
# 아래에 코드를 작성하세요.
````

````python
given = input()
if len(given) % 2:
    print(given[len(given) // 2])
else:
    print(given[len(given) // 2 - 1] + given[len(given) // 2])
````

# 소수 찾기

> 조건, 반복문을 응용하여 numbers 리스트의 요소들이 소수인지 아닌지 판단하는 코드를 작성하세요.

---

예시)

```python
numbers = [26, 39, 51, 53, 57, 79, 85]

26 는 소수가 아닙니다. 2 는 26 의 인수입니다.
39 는 소수가 아닙니다. 3 는 39 의 인수입니다.
51 는 소수가 아닙니다. 3 는 51 의 인수입니다.
53 는 소수입니다.
57 는 소수가 아닙니다. 3 는 57 의 인수입니다.
79 는 소수입니다.
85 는 소수가 아닙니다. 5 는 85 의 인수입니다.
```

````python
numbers = [26, 39, 51, 53, 57, 79, 85]

# 아래에 코드를 작성하세요.
````

````python
for each in numbers:
    isPrime = True
    for divider in range(2, int(each ** 0.5) + 1):
        if each % divider == 0:
            print(f'{each} 는 소수가 아닙니다. {divider} 는 {each} 의 인수입니다.')
            isPrime = False
    if isPrime:
        print(f'{each} 는 소수입니다.')
````
